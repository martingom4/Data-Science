```python
#3. Elimina los datos atípicos del dataset. Para las siguientes preguntas, considera el dataset "limpio"

# Definimos una función para eliminar los datos atípicos utilizando el rango intercuartílico (IQR)
def eliminar_outliers(df, columns):
    for column in columns:
        # Calcular los cuartiles y el rango intercuartílico (IQR)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Definir los límites para los datos atípicos
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR

        # Filtrar los datos dentro del rango de los límites
        df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]

    return df

# Seleccionamos las columnas numéricas del dataframe para eliminar atípicos
columnas_cuantitativas = ['age', 'BMI', 'glucose', 'totChol', 'sysBP', 'diaBP', 'heartRate', 'cigsPerDay']
df_limpio = eliminar_outliers(df, columnas_cuantitativas)
print(f"{df_limpio.shape[0]}")
# Mostramos la cantidad de datos restantes después de eliminar los atípicos
df_limpio.describe()
```

```python
#7. Ser hombre, ¿influye en el promedio de cigarrillos consumidos por día, dentro de los pacientes fumadores? Explica

import scipy.stats as stats

# Filtrar solo los pacientes fumadores y eliminar valores nulos en 'cigsPerDay'
fumadores = df[(df['currentSmoker'] == 'Yes') & df['cigsPerDay'].notnull()]

#Dividimos los datos en hombres y mujeres fumadores
hombres_fumadores = fumadores[fumadores['sex'] == 'M']['cigsPerDay']
mujeres_fumadoras = fumadores[fumadores['sex'] == 'F']['cigsPerDay']

#Calculamos el promedio de cigarrillos consumidos por hombres y mujeres fumadores
promedio_hombres = hombres_fumadores.mean()
promedio_mujeres = mujeres_fumadoras.mean()

print(f"Promedio de cigarrillos por día en hombres fumadores: {promedio_hombres}")
print(f"Promedio de cigarrillos por día en mujeres fumadoras: {promedio_mujeres}")

#Realizamos una prueba t de dos muestras para comparar los promedios
t_stat, p_valor = stats.ttest_ind(hombres_fumadores, mujeres_fumadoras, equal_var=False)

#Mostramos los resultados de la prueba t
print(f"Estadístico t: {t_stat}")
print(f"P-valor: {p_valor}")
```

```python
#Los hombres fumadores fuman en promedio unos 22 cigarrillos por día, mientras que las mujeres
#fumadoras fuman alrededor de 14 cigarrillos por día. Esto ya muestra una diferencia notable.
#Estadístico t: Nos ayuda a medir la diferencia entre los dos grupos (hombres y mujeres).
#Un valor más alto indica una mayor diferencia entre los promedios.

#P-valor: Nos dice si la diferencia observada es significativa o si podría haber ocurrido por casualidad.
#Un p-valor muy pequeño (como 4.22e-68) significa que es extremadamente improbable que la diferencia
#observada ocurra por casualidad.

#Por lo tanto, la respuesta es "Sí", ser hombre influye en el promedio de cigarrillos consumidos por día
#dentro de los pacientes fumadores. Los hombres, en promedio, fuman más cigarrillos por día que las mujeres.
#Y los números nos dicen que esta diferencia no es solo casualidad, sino que es muy significativa.
#Los datos nos muestran claramente que los hombres fumadores tienden a fumar más cigarrillos al día
#en comparación con las mujeres fumadoras.
```

```python
#8. Construye un modelo de regresión lineal que permita relacionar 6 variables del
#dataset con el índice de masa corporal. (Debes incluir al menos dos variables
#cualitativas). Evalúa tu modelo y explica.

#Importamos las librerías necesarias
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Variables seleccionadas para el modelo
variables_cuantitativas = ['age', 'heartRate', 'diaBP', 'BPMeds']
variables_cualitativas = ['sex', 'currentSmoker']

# Creamos un DataFrame con las variables seleccionadas
df_model = df[['BMI'] + variables_cuantitativas + variables_cualitativas]

# Codificamos las variables cualitativas usando dummy variables
df_model = pd.get_dummies(df_model, columns=variables_cualitativas, drop_first=True)

# Imputamos valores faltantes con la mediana en cada columna de las variables seleccionadas
df_model[variables_cuantitativas] = df_model[variables_cuantitativas].fillna(df_model[variables_cuantitativas].median())

# Eliminamos filas con valores nulos en la variable objetivo 'BMI'
df_model = df_model.dropna(subset=['BMI'])

# Dividimos los datos en variables independientes (X) y dependiente (y)
X = df_model.drop('BMI', axis=1)  # Variables independientes
y = df_model['BMI']  # Variable dependiente

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creamos y ajustamos el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Hacemos predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Calculamos las métricas de evaluación del modelo
mse = mean_squared_error(y_test, y_pred)  # Error Cuadrático Medio
r2 = r2_score(y_test, y_pred)  # Coeficiente de Determinación (R²)

# Revisamos los resultados de la evaluación del modelo
print(f"Error Cuadrático Medio (MSE): {mse}")
print(f"Coeficiente de Determinación (R²): {r2}")

# Coeficientes de las variables en el modelo
coeficientes = pd.DataFrame(model.coef_, X.columns, columns=['Coeficiente'])
print("\nCoeficientes de las variables en el modelo:")
print(coeficientes)

```

```python
#9. Construye un modelo de regresión logística para predecir el riesgo de sufrir una
#enfermedad coronaria en los próximos diez años, a partir de las variables descritas.
#Verifica el balanceo de datos y evalúa tu modelo.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report, roc_auc_score, roc_curve, average_precision_score, precision_recall_curve
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt

# Contar los valores en la columna TenYearCHD para ver si aplicamos balanceo
balanceo = df['TenYearCHD'].value_counts()
print("Balanceo = ", balanceo)

# Separar variables independientes X y dependiente y
X = df.drop(columns=["TenYearCHD"])
y = df['TenYearCHD']

# Convertir variables categóricas en binarias usando pd.get_dummies
X = pd.get_dummies(X, columns=['sex', 'currentSmoker'], drop_first=True)

# Imputar valores faltantes
imputer = SimpleImputer(strategy='median')
X = imputer.fit_transform(X)

# Aplicar SMOTE para balancear los datos
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Ajustar el modelo de regresión logística
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Predicciones y evaluación
y_pred = log_reg.predict(X_test)
y_proba = log_reg.predict_proba(X_test)[:, 1]  # Probabilidad para la clase positiva (1)

# Cálculo de métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Cálculo del AUC-ROC y AUC-PR
auc_roc = roc_auc_score(y_test, y_proba)
auc_pr = average_precision_score(y_test, y_proba)

print("Exactitud (accuracy):", accuracy)
print("Precisión (precision):", precision)
print("\nMatriz de Confusión:\n", conf_matrix)
print("\nReporte de Clasificación:\n", classification_rep)
print("\nAUC-ROC:", auc_roc)
print("AUC-PR:", auc_pr)

# Graficar la curva ROC
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(10, 5))
plt.plot(fpr, tpr, color='blue', label=f'Curva ROC (AUC = {auc_roc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.title("Curva ROC")
plt.legend()
plt.show()

# Graficar la curva Precision-Recall
precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_proba)
plt.figure(figsize=(10, 5))
plt.plot(recall_vals, precision_vals, color='purple', label=f'Curva Precision-Recall (AUC-PR = {auc_pr:.2f})')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Curva Precision-Recall")
plt.legend()
plt.show()
```

```python
#b) Modelo de Regresión Logistica:

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report, roc_auc_score, average_precision_score, roc_curve, precision_recall_curve
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import pandas as pd

# Variables independientes X y dependiente y
X = df.drop(columns=["TenYearCHD"])
y = df['TenYearCHD']

# Convertir variables categóricas en binarias usando pd.get_dummies
X = pd.get_dummies(X, columns=['sex', 'currentSmoker'], drop_first=True)

# Imputar valores faltantes
imputer = SimpleImputer(strategy='median')
X = imputer.fit_transform(X)

# Aplicar SMOTE para balancear los datos
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Convertir a DataFrame para facilitar la separación por fumadores y no fumadores
X_res_df = pd.DataFrame(X_res, columns=df.drop(columns=["TenYearCHD"]).columns)
X_res_df['TenYearCHD'] = y_res

# Separar por fumadores y no fumadores
df_fumador = X_res_df[X_res_df['currentSmoker'] == 1].drop(columns=['currentSmoker'])
df_no_fumador = X_res_df[X_res_df['currentSmoker'] == 0].drop(columns=['currentSmoker'])

# Función para entrenar, predecir y calcular métricas para cada grupo
def evaluate_model(X, y, title):
    # División en entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Entrenar modelo
    log_reg = LogisticRegression(max_iter=1000)
    log_reg.fit(X_train, y_train)

    # Predicciones y probabilidades
    y_pred = log_reg.predict(X_test)
    y_proba = log_reg.predict_proba(X_test)[:, 1]

    # Métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred)

    # AUC-ROC y AUC-PR
    auc_roc = roc_auc_score(y_test, y_proba)
    auc_pr = average_precision_score(y_test, y_proba)

    print(f"Resultados para {title}:")
    print(f"Exactitud (accuracy): {accuracy}")
    print(f"Precisión (precision): {precision}")
    print(f"\nMatriz de Confusión:\n{conf_matrix}")
    print(f"\nReporte de Clasificación:\n{classification_rep}")
    print(f"\nAUC-ROC: {auc_roc}")
    print(f"AUC-PR: {auc_pr}")

    # Graficar Curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.figure(figsize=(5, 2))
    plt.plot(fpr, tpr, label=f'{title} - ROC (AUC = {auc_roc:.2f})', color='blue')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel("Tasa de Falsos Positivos")
    plt.ylabel("Tasa de Verdaderos Positivos")
    plt.title(f"Curva ROC - {title}")
    plt.legend()
    plt.show()

    # Graficar Curva Precision-Recall
    precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_proba)
    plt.figure(figsize=(5, 2))
    plt.plot(recall_vals, precision_vals, label=f'{title} - Precision-Recall (AUC = {auc_pr:.2f})', color='purple')
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"Curva Precision-Recall - {title}")
    plt.legend()
    plt.show()

# Evaluar modelo para fumadores
evaluate_model(df_fumador.drop(columns='TenYearCHD'), df_fumador['TenYearCHD'], "Fumadores")

# Evaluar modelo para no fumadores
evaluate_model(df_no_fumador.drop(columns='TenYearCHD'), df_no_fumador['TenYearCHD'], "No Fumadores")
```
